```python
import numpy as np
import cv2 as cv

from matplotlib import pyplot as plt
```

```python
img = cv.imread('./images/cat.png', 0)
```

```python
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
hist, bins = np.histogram(img.flatten(), 256, [0,256])
```

```python
cumsum = hist.cumsum()
cumsum_normalized = cumsum / cumsum[-1]
```

```python
transf = (cumsum * 255) / (cumsum.max() - cumsum.min())
transf = transf.astype('uint8')

equ1 = transf[img]
plt.imshow(equ1, cmap='gray')
plt.show()
```

```python
equ2 = cv.equalizeHist(img)
plt.imshow(equ2, cmap='gray')
plt.show()
```
