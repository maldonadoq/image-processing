#### Vectorized Computation

```python
A = np.array([[0.84, -0.26, 0.92], [0.42, 0.31, -0.73]])

B = A * 10
print(B)

C = A + A
print(C)
```

To set all of the negative values in the array to 0:

```python
A[A < 0] = 0
print(A)
```

The numpy.where statement is a vectorized version of the ternary expression x if condition else y.

Example: take a value from xarray whenever the corresponding value in cond is True, otherwise take the value
from yarray. A possible solution might be:

```python
xarray = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarray = np.array([2.1, 2.2, 2.3, 2.4, 2.5])

cond = np.array([True, False, True, True, False])

result = [(x if c else y) for x, y, c in zip(xarray, yarray, cond)]

print(result)
```

This solution has some problems. It will not be fast for large arrays. Moreover, it will not work with multidimensional arrays. A m18 Images using NumPy and SciPy

```python
result = np.where(cond, xarray, yarray)
print(result)
```

To calculate and plot sine and cosine values of an array of angles given in radians:

```python
x = np.linspace(-np.pi, np.pi, 100)

plt.subplot(2,2,1)
plt.plot(x, np.sin(x))
plt.xlabel('angle [rad]')
plt.ylabel('sin(x)')

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.xlabel('angle [rad]')
plt.ylabel('cos(x)')

plt.axis('tight')
plt.subplot(2,1,2)
plt.plot(x, np.sin(x) * np.cos(x))
plt.title('sin(x)*cos(x)')
plt.tight_layout()

plt.show()
```

#### Images using NumPy and SciPy

```python
import imageio

# open image file and stores it in a numpy array
img = np.array(imageio.imread('images/iris.jpg'))

# print image dimensions and type
print(img.shape, img.dtype)

# show image
plt.imshow(img, cmap='gray')
plt.show()

# save image in JPG format
imageio.imsave('images/iris.jpg', img)

# calculate some statistical information
print(img.min(), img.mean(), img.max())

# apply rotation transformation
f = np.flipud(img)
plt.imshow(f)
plt.show()

# smooth image with Gaussian filter
g = ndimage.gaussian_filter(img, sigma=7)
h = ndimage.gaussian_filter(img, sigma=11)
plt.imshow(g)
plt.show()
plt.imshow(h)
plt.show()
```
